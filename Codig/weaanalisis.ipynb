```python
table_name = 'datos_lol'
df9.columns = df9.columns.str.replace('[^a-zA-Z0-9_]', '_')
try:
    # Crear una conexión a MySQL usando mysql.connector
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password
    )

    if connection.is_connected():
        cursor = connection.cursor()

        # Crear la base de datos si no existe
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
        print(f"Base de datos '{database_name}' creada o ya existente.")

        # Seleccionar la base de datos
        cursor.execute(f"USE {database_name}")

        # Obtener nombres de columna limpios y tipos de datos
        column_names = df9.columns.tolist()
        cleaned_column_names = [name.replace(' ', '_') for name in column_names]  # Reemplazar espacios con guiones bajos
        column_types = [df9[col].dtype for col in column_names]

        # Crear la consulta CREATE TABLE
        create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ("
        for i in range(len(cleaned_column_names)):
            column_name = cleaned_column_names[i]
            column_type = str(column_types[i])
            if "int" in column_type:
                create_table_query += f"`{column_name}` INT"  # Utiliza comillas inversas para escapar el nombre
            elif "float" in column_type:
                create_table_query += f"`{column_name}` FLOAT"  # Utiliza comillas inversas para escapar el nombre
            else:
                create_table_query += f"`{column_name}` VARCHAR(255)"  # Utiliza comillas inversas para escapar el nombre
            if i < len(cleaned_column_names) - 1:
                create_table_query += ", "
        create_table_query += ");"

        # Ejecutar la consulta para crear la tabla
        cursor.execute(create_table_query)
        print(f"Tabla '{table_name}' creada exitosamente.")

except Error as e:
    print("Error al conectarse a la base de datos:", e)

finally:
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
        print("Conexión a MySQL cerrada.")
```
